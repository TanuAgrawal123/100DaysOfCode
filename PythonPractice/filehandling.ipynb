### open function

```python
f=open('text.txt','r') #default mode
#open(filename, mode). 
print(f.read())
```

```python
f.close()
```

### opening the file with context manager

```python
with open('text.txt', 'r+') as f:
    print(f.read())
    
```

```python
f.read()#throws an error as it automatically close the file
```

```python
with open('text.txt', 'r+') as f:
    print(f.tell())# current cursor position
    
    
   
   
    print(f.readline(),end='')# to read line by line
    print(f.readline(),end='')
    print(f.readline(),end='')
    
```

```python
with open('text.txt', 'r+') as f:# for large file 
    for line in f:
        print(line)
    
    
```

##### writing in file

```python
with open('text1.txt','w') as f: # if file does not exist then it make new file otherwise overwrite existing file
    f.write('tanu')
    f.write('hello')
    #now content of file be tanu hello
    f.seek(0)
    #set cursor at starting position
    f.write('anubha')# overwrite previous content(anubhallo)
   
```

#### copying the content of the file

```python
with open('text.txt','r') as rf:
    with open('text_copy.txt','w') as wf:
        for line in rf:
            wf.write(line)
```

### copying a pictiure file by switching to binary mode


```python
with open('pic.png','rb') as rf:
    with open('pic_copy.png','wb') as wf:
        for line in rf:
            print(wf.write(line))
        
```
